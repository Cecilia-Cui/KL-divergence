```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('kl_divergence_results.csv')

plt.figure(figsize=(14, 8)) 
plt.plot(data['folder'], data['kl_12'], label='KL 12', marker='o')
plt.plot(data['folder'], data['kl_13'], label='KL 13', marker='s')
plt.plot(data['folder'], data['kl_23'], label='KL 23', marker='^')

plt.title('Fig.1 KL Divergence')
plt.xlabel('Index') 
plt.ylabel('KL Divergence')
plt.legend()
plt.grid(True)

plt.xticks(ticks=data['folder'][::10], rotation=45, fontsize=10)  

plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# KL 12 Distribution
plt.subplot(3, 1, 1)
plt.hist(data['kl_12'], bins=30, color='blue', alpha=0.7)
plt.title('KL 12 Distribution')
plt.xlabel('KL 12 Values')  
plt.ylabel('Frequency') 

# KL 13 Distribution
plt.subplot(3, 1, 2)
plt.hist(data['kl_13'], bins=30, color='green', alpha=0.7)
plt.title('KL 13 Distribution')
plt.xlabel('KL 13 Values')  
plt.ylabel('Frequency')  

# KL 23 Distribution
plt.subplot(3, 1, 3)
plt.hist(data['kl_23'], bins=30, color='red', alpha=0.7)
plt.title('KL 23 Distribution')
plt.xlabel('KL 23 Values')  
plt.ylabel('Frequency') 

plt.suptitle('Fig.2 KL Divergence Distribution', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # 
plt.show()
```
