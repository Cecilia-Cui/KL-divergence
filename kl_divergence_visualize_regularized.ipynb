```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取KL散度结果的CSV文件
kl_results_path = 'kl_divergence_results_regularized.csv'  # 更新为您的实际文件路径
kl_df = pd.read_csv(kl_results_path)

# 生成KL散度的折线图
def plot_kl_divergence(dataframe, output_file='kl_divergence_plot.png'):
    plt.figure(figsize=(14, 7))
    
    # 绘制每个KL散度的折线图
    for col in ['kl_12', 'kl_13', 'kl_23']:
        plt.plot(dataframe['group'], dataframe[col], marker='o', label=col)

    # 设置图表属性
    plt.title('KL Divergence Comparison Across Groups')
    plt.xlabel('Group')
    plt.ylabel('KL Divergence')
    plt.xticks([])  # 不显示x轴的坐标标注
    plt.grid(True)
    plt.legend()
    
    # 保存并显示图表
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()

# 调用函数生成图表
plot_kl_divergence(kl_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取KL散度结果的CSV文件
kl_results_path = 'kl_divergence_results_regularized.csv'  # 更新为您的实际文件路径
kl_df = pd.read_csv(kl_results_path)

# 创建直方图来展示 KL 散度的分布
plt.figure(figsize=(12, 8))

# KL 12 Distribution
plt.subplot(3, 1, 1)
plt.hist(kl_df['kl_12'].dropna(), bins=30, color='blue', alpha=0.7)
plt.title('KL 12 Distribution')
plt.xlabel('KL 12 Values')
plt.ylabel('Frequency')

# KL 13 Distribution
plt.subplot(3, 1, 2)
plt.hist(kl_df['kl_13'].dropna(), bins=30, color='green', alpha=0.7)
plt.title('KL 13 Distribution')
plt.xlabel('KL 13 Values')
plt.ylabel('Frequency')

# KL 23 Distribution
plt.subplot(3, 1, 3)
plt.hist(kl_df['kl_23'].dropna(), bins=30, color='red', alpha=0.7)
plt.title('KL 23 Distribution')
plt.xlabel('KL 23 Values')
plt.ylabel('Frequency')

# 设置总标题
plt.suptitle('KL Divergence Distribution', fontsize=16)

# 调整子图布局以防止标题重叠
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
